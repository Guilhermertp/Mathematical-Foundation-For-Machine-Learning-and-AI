```python
import sys
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

print('Python:{}'.format(sys.version))
print('Numpy:{}'.format(np.__version__))
print('Matplotlib:{}'.format(matplotlib.__version__))
```

```python
# generate 2D meshgrid
nx, ny = (100,100) #number of points we want in the meshgrid

x = np.linspace(0, 10, nx)
y = np.linspace(0, 10, ny)

xv, yv = np.meshgrid(x,y)

xv.shape # check the shape that should be 100 x 100
```

```python
#define a function to plot
def f(x,y):
    return x * (y**2)

# calculate Z value for each x,y point
z = f(xv, yv)
```

```python
# make a color plot to display the data
plt.figure(figsize=(14,12))
plt.pcolor(xv,yv,z) #Pcolor allows  to generate 2D image-style plots
plt.title('2D Color Plot of f(x,y) = xy^2')
plt.colorbar() # Add a colorbar to a plot.
plt.show()
```

```python
# generate 2D meshgrid for the gradient
nx, ny = (10,10)
x = np.linspace(0,10,nx)
y = np.linspace(0,10,ny)
xg, yg = np.meshgrid(x,y)

#calculate the gradient of f(x,y)
Gy, Gx = np.gradient(f(xg, yg)) #because numpy returns first rows then columns in the context of x and y, the rows are going to be y and the columns x
# numpy will deal first with the y values the rows and then the x values the columns, important aspect to keep in mind
```

```python
# make a color plot to display the data
plt.figure(figsize=(14,12))
plt.pcolor(xv,yv,z) #Pcolor allows  to generate 2D image-style plots
plt.title('2D Color Plot of f(x,y) = xy^2')
plt.colorbar() # Add a colorbar to a plot.
plt.quiver(xg, yg, Gx, Gy, scale = 1000, color = 'w')
plt.show()
```

```python
#to verifify if the plot above is correct, if numpy was calculating the gradient correctly

#double check by ourselfs finding the partial derivatives of x and y
#calculate the gradient of f(x,y) = xy^2

#function for partial derivative of x
def ddx(x,y):
    return y ** 2

#function for partial derivative of y
def ddy(x,y):
    return 2 * x * y

#x component of the gradient
Gx = ddx(xg, yg)
#y component of the gradient
Gy = ddy(xg, yg)

# we are goin to verify comparing plots

# make a color plot to display the data
plt.figure(figsize=(14,12))
plt.pcolor(xv,yv,z) #Pcolor allows  to generate 2D image-style plots
plt.title('2D Color Plot of [y^2, 2xy]')
plt.colorbar() # Add a colorbar to a plot.
plt.quiver(xg, yg, Gx, Gy, scale = 1000, color = 'w')
plt.show()
```

```python
# As expected the plots are the same, it verifys that the gradient was calculated correctly :) !!!
```
